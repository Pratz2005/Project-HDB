<h2>Data cleaning and extrapolation for merging into main data</h2>
<h3> Used extrapulation to get the values before 2002 and after 2017,and interpolation to get the values between the given 4 Years  </h3>

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from scipy.interpolate import interp1d

df_costExtrapo = pd.DataFrame(data_HouseholdExpend)
# Set 'Year' as index
df_costExtrapo.set_index('Year', inplace=True)

# Extrapolate the values for each year from 1990 to 2021 using interpolation
years_range = list(range(1990, 2022))
extrapolated_data = {}
for column in df_costExtrapo.columns:
    # Create an interpolation function for each column
    f = interp1d(df_costExtrapo.index, df_costExtrapo[column], kind='linear', fill_value='extrapolate')
    # Use the interpolation function to predict values for the desired years
    extrapolated_data[column] = f(years_range)

# Update the predicted values for the years 2002, 2007, 2012, and 2017 to match the provided values
for year in [2002, 2007, 2012, 2017]:
    for column in df_costExtrapo.columns:
        extrapolated_data[column][years_range.index(year)] = df_costExtrapo.loc[year, column]

# Create a dataframe for extrapolated data
df_extrapolated = pd.DataFrame(extrapolated_data, index=years_range)

# Display the extrapolated dataframe
print(df_extrapolated)
```

<h2>Reorganise df_extrapolated into a format matching main data</h2>

```python
df_FullExpense = pd.DataFrame(df_extrapolated)
# Reset index to make 'Year' a column again

df_FullExpense.reset_index(drop=False, inplace=True)
# Create new columns for each room type
df_FullExpense['Year'] = df_FullExpense['index']
df_FullExpense['1 ROOM'] = df_FullExpense['1- & 2-Room Flats 2/']
df_FullExpense['2 ROOM'] = df_FullExpense['1- & 2-Room Flats 2/']
df_FullExpense['3 ROOM'] = df_FullExpense['3-Room Flats']
df_FullExpense['4 ROOM'] = df_FullExpense['4-Room Flats']
df_FullExpense['5 ROOM'] = df_FullExpense['5-Room & Executive Flats']
df_FullExpense['EXECUTIVE'] = df_FullExpense['5-Room & Executive Flats']
df_FullExpense['MULTI GENERATION'] = df_FullExpense['5-Room & Executive Flats']

# Drop the original columns
df_FullExpense.drop(['index','1- & 2-Room Flats 2/', '3-Room Flats', '4-Room Flats', '5-Room & Executive Flats'], axis=1, inplace=True)

# Reorder the columns according to the new layout
#df_FullExpense = df_FullExpense[['Year', '1 ROOM', '2 ROOM', '3 ROOM', '4 ROOM', '5 ROOM', 'EXECUTIVE', 'MULTI GENERATION']]
```

<H2>Merge expenditure into main dataset</H2>

```python
# Merge the main dataframe with the labeled dataframe on the 'Year' column
Merged_RPI_Expense_df = data_RPI_merge.merge(df_FullExpense, left_on='year', right_on='Year', how='left')

# Function to fill in expenditure value based on the type of home
def fill_expenditure(row):
    home_type = row['flat_type'] 
    if home_type == '1 ROOM':
        return row['1 ROOM']
    elif home_type == '2 ROOM':
        return row['2 ROOM']
    elif home_type == '3 ROOM':
        return row['3 ROOM']
    elif home_type == '4 ROOM':
        return row['4 ROOM']
    elif home_type == '5 ROOM':
        return row['5 ROOM']
    elif home_type == 'EXECUTIVE':
        return row['EXECUTIVE']
    elif home_type == 'MULTI GENERATION':
        return row['MULTI GENERATION']
    else:
        return None  # Handle cases where 'Type_of_Home' doesn't match any of the specified types

# Apply the function to create a new column 'Expenditure' in the main dataframe
Merged_RPI_Expense_df['Expenditure'] = Merged_RPI_Expense_df.apply(fill_expenditure, axis=1)
Merged_RPI_Expense_df.drop(['Year','1 ROOM', '2 ROOM', '3 ROOM', '4 ROOM','5 ROOM','EXECUTIVE','MULTI GENERATION'], axis=1, inplace=True)
```

<h2>Explore a more through version of classification model: Random Forest classifier</h2>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
#use same test train set as decision tree
#however y needs to be encoded in order to fit randomforestclassifier



columns_of_interest = ['year', 'price_psm cpi_adj','area_sqm','Expenditure','flat_type']
classification_df=Merged_RPI_Expense_df[columns_of_interest]

X = classification_df.drop('flat_type', axis=1)  # drop target variable
#y = classification_df['flat_type']

#1. Encode Y for random forest
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(classification_df['flat_type'])

# 2. Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# 3. Model Selection
rf_classifier = RandomForestClassifier(n_estimators=15,max_depth=4, random_state=33)

# 4. Model Training
rf_classifier.fit(X_train, y_train)

# 5. Model Evaluation
accuracy = rf_classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```
